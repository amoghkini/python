Rules of dictionary
1. Dictionary has no indexing
2. Dictionary is a mutable object
3. Keys -> immutable, values - mutable
3. keys should be unique

1. Mutable datatypes - list, sets,dictionary
2. immutable datatypes - string, tuple, int, float, boolean, complex
