### Problems on list

"what is your name".title() ---> implement this from scratch

```python
a = "what is your name"
l = a.split()
n = list()
for i in l:
    n.append(i.capitalize())
    
c = " ".join(n)
print(c)
```

"abc@gmail.com" - find data before @

```python
l = "abc@gmail.com"

print(l[:l.find("@")])
```

```python
l = [1,1,1,2,2,3,3,4,5] # find unique elements
m = []

for i in l:
    if i not in m:
        m.append(i)
        
print(m)
```
