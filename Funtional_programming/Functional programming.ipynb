### Higher order functions
##### functions which requires function as input or functino which returns function as output

```python
# Find
# 1. Sum of even numbers
# 2. Sum of odd numbers
# 3. Sum of numbers divisible by 3
l = [11,14,21,54,45,12,89,48,65,76]

```

```python
def find_sum(l):
    sum_even = 0
    sum_odd = 0
    sum_div3 = 0
    for i in l:
        if i%2 == 0:
            sum_even += i
        if i%2 != 0:
            sum_odd += i
            
        if i%3 == 0:
            sum_div3 += i
                
    return (sum_even,sum_odd,sum_div3)

for i in find_sum(l):
    print(i)
    
```

#### Now we will implement the same logic using lambda functinos and higher order functions(Which takes function as input)

```python
def find_sum_hof(func,l):
    total = 0
    
    for i in l:
        if func(i):
            total += i
    return total

x = lambda x:x%2 == 0
y = lambda x:x%2 != 0
z = lambda x:x%3 == 0

print(find_sum_hof(x,l))
print(find_sum_hof(y,l))
print(find_sum_hof(z,l))
```

### Python built Higher Order Functions
1. Map

### 1. Map() - We can pass a function and iterable to the map() function and it will map the iterable with given function.

```python
l = [1,2,3,4,5,6,7,8,9]

map(lambda x:x*2,l)
```

#### Map function will return a object hence we need to typecast it to a list

```python
list(map(lambda x:x*2,l))
```

```python
list(map(lambda x:x%2==0,l))
```

### 2. Filter() - filter fitlers the data based on given conditions whereas map performs operation on each item of given iterable.

```python
list(filter(lambda x:x>4,l))
```

```python
list(map(lambda x:x>4,l))
### This is the difference between map and filer. map func returns each element of iterable
```

```python
fruits = ['apple','orange','mango','guava']
```

```python
list(filter(lambda x: 'e' in x,fruits))
```
