#### 1. Sets do not have duplicates
#### 2. Sets have no indexing/slicing
#### 3. Sets don't allow mutable data types
#### 4. Set itself is a mutable data type
    


### 1.Create set

```python
s1 = {} # This will create empty dictionary rather than empty set
s1
```

```python
type(s1)
```

```python
s1 = set()
s1
```

```python
s1 = {1,2,3,4,5,6,7}
```

```python
s2 = {"hello",4.2,4,True}
```

```python
#Rule 1
s3 =  {1,2,2,2,3,4,5}
s3
```

```python
# Rule 3
s4 = {[1,2,3,4],"string"}
s4
```

```python
s5 = {(1,2,3,4),"Hello"}
```

```python
s5
```

```python
### Sets does not follow indexing/slicing given by as. It uses hashing.
```

### Access items

```python
## Rule 2
s3[0]
```

```python
s3[0:2]
```
