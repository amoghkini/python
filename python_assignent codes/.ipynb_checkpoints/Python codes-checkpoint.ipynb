## Question 2

Write a code to check whether a given number is a palindrome.

```python
number = int(input("Enter number to check if it is palindrome:"))
number_backup = number
number_after = 0
while number:

    rem = number % 10
    number = number // 10
    number_after = number_after * 10 + rem

if number_after == number_backup:
    print("The given number is palindrome")
else:
    print("The given number is not a palindrome")
```

## Question 4
A three digit number is said to be an “Armstrong number” if the sum of the third power of its individual digits is equal to the number itself.Write a program to check whether a number is armstrong or not. 

```python
number = int(input("Enter the three digit number to check if it is Armstrong number"))

if number < 100 or number > 999:
    print("Please enter 3 digit number")
else:
    number_bkp = number
    cube_sum = 0
    while number:
        cube = pow(number % 10,3)
        cube_sum = cube_sum  + cube
        number = number // 10
    if number_bkp == cube_sum:
        print(number_bkp, " is armstrong number")
    else:
        print(number_bkp, " is not a armstrong number")
```

If there are 500 students who have joined the university, write a code to calculate and display the final fees to be paid by each student.

Scholarship amount=course fee * (scholarship%)
Final fee= course fee - scholarship amount



```python
import sys
def calculateFees(branch,score,fees):
    if branch == 1:
        if score >=90 :
            scholar_amount = fees * 50/100
        elif score % 2 != 0:
            scholar_amount = fees * 5/100
    else:
        if score >=85 :
            scholar_amount = fees * 50/100
        elif score % 7 != 0:
            scholar_amount = fees * 5/100
    
    return fees - scholar_amount

print("Enter branch of study:")
print("1:Arts \n2:Engineering")
print("-----------------------")
branch = int(input())
if branch != (1 or 2):
    print("Invalid branch")
    sys.exit(0)
score = int(input("Enter your score:"))
fees = int(input("Enter the course fees:"))

print("The final course fee is ", calculateFees(branch,score,fees))
```

## Question 6

The flight ticket rates for a round-trip (Mumbai->Dubai) were as follows:
* Rate per Adult: Rs. 37550.0
* Rate per Child: 1/3rd of the rate per adult
* Service Tax: 7% of the ticket amount (including all passengers)
* As it was a holiday season, the airline also offered a 10% discount on the final ticket cost (after inclusion of the service tax).
* Find and display the total ticket cost for a group which had adults and children.


```python
def price(adult_price,service_tax,discount,no_of_adults,no_of_childs):
    final_price_per_adult = adult_price + (adult_price * service_tax / 100)
    rate_per_child = 1/3
    child_price = rate_per_child * adult_price
    final_price_per_child =child_price + (child_price * service_tax / 100)
    
    total = ( final_price_per_adult * no_of_adults ) + (final_price_per_child * no_of_childs)
    
    amount = total - ( total * discount /100 )
    return amount
    
    
adult_price  = float(input("Enter the rate per adult:"))
service_tax  = float(input("Enter the service tax:"))
discount     = float(input("Enter the discount:"))
no_of_adults = int(input("Enter no of adults:"))
no_of_childs = int(input("Enter no of childs:"))

total = price(adult_price,service_tax,discount,no_of_adults,no_of_childs)
print("The total ticket cost is ",total)
```

## Question 9
Write a python program to find and display the product of three positive integer values based on the rule mentioned below:
 
It should display the product of the three values except when one of the integer value is 7. In that case, 7 should not be included in the product and the values to its left also should not be included.
If there is only one value to be considered, display that value itself. If no values can be included in the product, display -1.
 
Note: Assume that if 7 is one of the positive integer values, then it will occur only once. Refer the sample I/O given below

| Sample Input  |  Expected Output     |
| :---                        |       ---: |
| 1, 5, 3         |15           |
| 3, 7, 8           |  8         |
|7, 4, 3            | 12          |
|1, 5, 7             |  -1         |

```python
def product(num1,num2,num3,number):
    if num3 == number:
        return -1
    elif num2 == number:
        return num3
    elif num1 == number:
        return num2 * num3
    else:
        return num1 * num2 * num3
    
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
num3 = int(input("Enter third number:"))
number = int(input("Enter the number to be excluded:"))

if num1 < 1 or num2 < 1 or num3 < 1:
    print("One of the entered input is invalid")
else:
    ans = product(num1,num2,num3,number)
    print("The answer is:",ans)
```

## Question 11
 Write a Python program to generate the next 15 leap years starting from a given year. Populate the leap years into a list and display the list.


```python
def findLeapYear(year):
    if (year % 400 == 0) and (year % 100 == 0):
        return  True
    elif (year % 4 ==0) and (year % 100 != 0):
        return  True
    else:
        return  False
    
def nextFifteenLeapYear(year):
    lst = []
    while True:
        if findLeapYear(year):
            next_leap_year = year
            break
        else:
            year = year + 1
            findLeapYear(year)
            
    for i in range(1,16):
        lst.append(next_leap_year)
        next_leap_year = next_leap_year + 4
    return lst

year = int(input("Enter starting year:"))

ans = nextFifteenLeapYear(year)

for i in ans:
    print(i,"\t")
```

## Question 17
Represent a small bilingual (English-Swedish) glossary given below as a Python dictionary

{"merry":"god", "christmas":"jul", "and":"och", "happy":"gott", "new":"nytt", "year":"ar"} 

and use it to translate your Christmas wishes from English into Swedish.

That is, write a python function translate() that accepts the bilingual dictionary and a list of English words (your Christmas wish) and returns a list of equivalent Swedish words.


```python
def translateWish(dictionary,wish):
    return_wish = []
    for i in wish:
        i = i.lower()
        return_wish.append("" if dictionary.get(i) == None else dictionary.get(i))
        
    return return_wish
    
        
dictionary1 = {
    "merry":"god",
    "christmas":"jul",
    "and":"och",
    "happy":"gott",
    "new":"nytt",
    "year":"ar"
}

wish = ['merry',"christmas"]

print(translateWish(dictionary1,wish))
```

## Question 18
The road transport corporation (RTC) of a city wants to know whether a particular bus-route is running on profit or loss.


Assume that the following information are given: <br>
Price per litre of fuel = 70 <br>
Mileage of the bus in km/litre of fuel = 10 <br>
Price(Rs) per ticket = 80 <br>

The bus runs on multiple routes having different distance in kms and number of passengers.
Write a function to calculate and return the profit earned (Rs) in each route. Return -1 in case of loss.


```python
def  profit(Price_per_litre_of_fuel,Mileage_of_bus_in_km_per_litre_of_fuel,Price_per_ticket,distance,no_of_passengers):
    revenue = Price_per_ticket * no_of_passengers
    expenses = Price_per_litre_of_fuel*distance/Mileage_of_bus_in_km_per_litre_of_fuel
    return  revenue - expenses

Price_per_litre_of_fuel = 70
Mileage_of_bus_in_km_per_litre_of_fuel = 10
Price_per_ticket = 80

distance = 100
no_of_passengers = 12
print("The profit of company is ",profit(Price_per_litre_of_fuel,Mileage_of_bus_in_km_per_litre_of_fuel,Price_per_ticket,distance,no_of_passengers))
```

## Question 19
Given a string containing uppercase characters (A-Z), compress the string using Run Length encoding. Repetition of character has to be replaced by storing the length of that run.

Write a python function which performs the run length encoding for a given String and returns the run length encoded String.

Provide different String values and test your program.

Sample Input : AAAABBBBCCCCCCCC <br>
Expected Output: 4A4B8C

Sample Input: AABCCA  <br>
Expected Output: 2A1B2C1A

```python
def encodestring(string):
    save_char = string[0]
    count = 1
    encodestring = ""
    for i in range(1,len(string)):
        if save_char == string[i]:
            count +=1
        else:
            encodestring = encodestring+str(count)+save_char
            count=1
            save_char = string[i]
            
    encodestring = encodestring+str(count)+save_char
    
    
    return encodestring
    
string = "AABCCA"

print("Encoded string is ",encodestring(string))
```

## Question 20
Write a function, check_palindrome() to check whether the given string is a palindrome or not. The function should return true if it is a palindrome else it should return false.

Note: Initialize the string with various values and test your program. Assume that all the letters in the given string are all of the same case. Example: MAN, civic, WOW etc. 


```python
def check_palindrome(string1):
    string1 = string1.upper()
    string_rev = ""
    
    for i in reversed(range(len(string1))):
        string_rev = string_rev + string1[i]
        
    if string_rev == string1:
        return True
    return False

string1 = "WOW"
if check_palindrome(string1):
    print("Palindrome")
else:
    print("Not palindrome")
```

## Question 21
Care hospital wants to know the medical speciality visited by the maximum number of patients. Assume that the patient id of the patient along with the medical speciality visited by the patient is stored in a list. The details of the medical specialities are stored in a dictionary as follows: <br>
{
"P":"Pediatrics",
"O":"Orthopedics",
"E":"ENT
}

<br>
Write a function to find the medical speciality visited by the maximum number of patients and return the name of the speciality.

Note:
1.	Assume that there is always only one medical speciality which is visited by maximum number of patients.
2.	Perform case sensitive string comparison wherever necessary.


```python
inp = ["101","P","102","O","302","P","305","P"]
inp = ["101","O","102","O","302","P","305","E","401","O","656","O"]

dictionay = {"P":"Pediatrics", "O":"Orthopedics", "E":"ENT" }

def medicalspecility(dictionay,inp):
    dict_count = {}
    for i in inp:
        if i.isdigit():
            continue
        else:
            if dict_count.get(i) == None:
                dict_count[i] = 1
            else:
                dict_count[i] = dict_count[i] + 1
            
    return dict_count  

return_dict = medicalspecility(dictionay,inp)
max = 0
char = ""
for i,j in return_dict.items():
    if j > max:
        char = i
        max = j

print("The maximum treatment is taken for",dictionay.get(char))
```
