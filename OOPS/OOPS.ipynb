#### Set class does not contains __add__ magic method hence we are not able to perform addition of two sets. Similart, our class does not have add method thus system is throwing the error

```python
class Fraction:
    
    def __init__(self,n,d):
        self.num = n
        self.den = d
        
    def __str__(self):
        return "{}/{}".format(self.num,self.den)
    
    def __add__(self,other):  # self will contain 1st object and other will contain 2nd object
        temp_num = self.num * other.den + self.den + other.num
        temp_deno = self.den * other.den
        return "{}/{}".format(temp_num,temp_deno)
    
    def __sub__(self,other):  # self will contain 1st object and other will contain 2nd object
        temp_num = self.num * other.den - self.den + other.num
        temp_deno = self.den * other.den
        return "{}/{}".format(temp_num,temp_deno)
    
    def __mul__(self,other):  # self will contain 1st object and other will contain 2nd object
        temp_num = self.num * other.num
        temp_deno = self.den * other.den
        return "{}/{}".format(temp_num,temp_deno)
    
    def __truediv__(self,other):
        temp_num = self.num * other.den
        temp_deno = self.den * other.num
        return "{}/{}".format(temp_num,temp_deno)
```

```python
a = Fraction(5,6)
b = Fraction(8,9)
```

now by using class nae only we can access counter variable. But it's better to create getter and setter methods inside the class to access the static data. We can use @staticmethod to create static methods

```python
class Customer:
    
    ## Class/static variable
    __counter = 1
    
    def __init__(self,name,age):
        ## Instance variable
        self.name = name
        self.age = age
        self.sno = Customer.__counter
        Customer.__counter += 1
    
    @staticmethod
    def get_counter():        # For static methods, we do not pass self as object name is not required.
        return Customer.__counter
    
    @staticmethod
    def set_counter(new_counter):    # For static methods, we do not pass self as object name is not required.
        if type(new_counter) == str:
            print("Invalid counter")
        else:
            Customer.__counter = new_counter
            print("Counter changed successfully")
            
    def myinfo():
        return "My name is {} and id is {}. My age is {}.".format(self.name,self.sno,self.age)
```

### 2. Multilevel Inheritance

```python
class Product:
    def review(self):
        print("Parent class review function")
        
class Phone(Product):
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.price = price
        self.brand = brand
        self.camera = camera
        
    def buy(self):
        print("Buying a phone from Phone class")
        
class Smartphone(Phone):
    pass

s = Smartphone(25000,"Samsung",32)
s.buy()
s.review()
p = Phone(30000,"Apple",12)
p.review()
```

### 3. Hierarchical Inheritance

One parent and multiple children

```python
class Phone():
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.price = price
        self.brand = brand
        self.camera = camera
        
    def buy(self):
        print("Buying a phone from Phone class")
        
class Smartphone(Phone):
    pass

class FeaturePhone(Phone):
    pass

s = Smartphone(25000,"Samsung",32)
s.buy()
p = FeaturePhone(30000,"Apple",12)
p.buy()
```

### 4.Multiple Inheritance
multiple parents and one child

```python
class Phone:
    def __init__(self,price,brand,camera):
        print("Inside phone constructor")
        self.price = price
        self.brand = brand
        self.camera = camera
        
    def buy(self):
        print("Buying a phone from Phone class")

class Product:
    def review(self):
        print("Parent class review function")
        
    def buy(self):
        print("Buying a phone from Product class")
        
        
class Smartphone(Product,Phone):   # Here we can add n no of classes
    pass
'''As constructor is not created in Smartphone, it will try to invoke the constructor based on Classes inherited 
eg. it will first check Phone class for constructor, if it is not present then it will go for Product class'''

s = Smartphone(25000,"Samsung",32)
s.buy()
s.review()


```

### MRO: Method reservation order
In above example, buy() method is present in both product and phone class, hence system will execute based on othe order of class inherited.
